```python
from langchain.chat_models import init_chat_model

model = init_chat_model("openai/gpt-oss-120b", model_provider="groq")
```

```python
# invoking to build a converstation style call

from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage("You are a helpful assistant. Answer the user query to the best of your memory."),
    HumanMessage("Give me a very hard question to test the IQ of a person. I'll later ask you to find the iq based upon the answer the person gives" )
]

response = model.invoke(messages)
```

```python
# Lets test another model if they can solve this problem

from langchain.chat_models import init_chat_model

model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

response = model.invoke("""Answer the following question carefully, you'll be judged based upon your answer.
There are five houses in a row, each painted a different color: blue, green, red, white, and yellow. Each house is occupied by a person of a different nationality: American, British, Canadian, Indian, and Japanese. Each person has a different favorite drink: coffee, tea, milk, soda, and water. Using the following clues, can you determine the color of each house, the nationality of its occupant, and their favorite drink?

1. The Canadian lives in the first house.
2. The person who drinks milk lives next to the person who owns the yellow house.
3. The person who owns the yellow house drinks soda.
4. The person who drinks coffee lives in the house next to the British person.
5. The American lives in the red house.
6. The person who drinks tea lives in the house next to the person who drinks water.
7. The green house is to the immediate left of the white house.

What are the colors of the houses, the nationalities of their occupants, and their favorite drinks?""")

print(response.content)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage("You are a helpful assistant. Answer the user query to the best of your memory."),
    HumanMessage("""Tell the iq of the person based upon the answer to the *Five House Problem*.Just output one number which will be the iq.
   Answer:
    1. **The Canadian lives in the first house.**
   - This gives us a starting point. The first house is occupied by a Canadian, but we don't know the color of the house or their favorite drink yet.

2. **The person who drinks milk lives next to the person who owns the yellow house.**
   - This means the yellow house cannot be the first or the last house since the person who drinks milk needs to live next to it. Possible positions for the yellow house are 2, 3, or 4.

3. **The person who owns the yellow house drinks soda.**
   - Now we know the favorite drink of the person in the yellow house, but not their nationality or the exact position of the house.

4. **The person who drinks coffee lives in the house next to the British person.**
   - This clue indicates the British person and the coffee drinker are in adjacent houses, but we don't know their positions or the colors of their houses.

5. **The American lives in the red house.**
   - This gives us another fixed point: the American lives in the red house, but we don't know the position of the red house or the favorite drink of the American.

6. **The person who drinks tea lives in the house next to the person who drinks water.**
   - Similar to clue 4, this indicates the tea drinker and the water drinker are in adjacent houses.

7. **The green house is to the immediate left of the white house.**
   - This clue helps with the arrangement of the houses. The green and white houses are together, in that order.

Given these clues, let's try to arrange the houses:
...
4. **Green - Indian - Tea**
5. **White - Japanese - Water**""")
]

response = model.invoke(messages)
```

```python
print(response.content)
```
